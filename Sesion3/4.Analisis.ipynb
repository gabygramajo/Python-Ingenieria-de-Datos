¿Cuales son es el productos mas caros y mas baratos?

```python
# El producto más caro
index = df_products['product_price'].idxmax()

df_products.loc[index]
```

```python
# El producto menos caro
index = df_products['product_price'].idxmin()

df_products.loc[index]
```

¿Cuales son los 5 productos mas caros y mas baratos?

```python
# mas caros
df_products.nlargest(5, 'product_price')
```

```python
# mas baratos
df_products.nsmallest(5, 'product_price')
```
