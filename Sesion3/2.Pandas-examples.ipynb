```python
# Detectar nulos con True o False
df_sales.isnull()
```

```python
# Sumar los nulos por columna
df_sales.isnull().sum()
```

```python
# Sumar los nulos por columna y ordenarlos desc
df_sales.isnull().sum().sort_values(ascending=False) 
```

```python
# Sacar porcentaje de nulos
df_sales.isnull().sum() / df_sales.isnull().count()
```

```python
# Seleccionar columnas segun el tipo de dato
df_sales.select_dtypes(include=['float'])
```

```python
# Seleccionar columnas específicas por nombre
df_sales[['CUSTOMERNAME', 'STATE']]
```

```python
# Slacing con columnas
df_sales[df_sales.columns[2:5]]
```

```python
# Slacing con filas

"""
Se usa para acceder a los datos mediante etiquetas (nombres de filas y columnas).

Referencia: Usa los nombres específicos que ves en el índice y en las columnas.

Slicing: A diferencia de las listas normales de Python, el rango en loc es inclusivo (incluye tanto el inicio como el final).
"""
df_sales.loc[10:20, ['CUSTOMERNAME', 'STATE']]
```

```python
# Slacing mediante indice
"""
Se usa para acceder a los datos mediante posiciones enteras (índices numéricos de 0 a $n-1$).Referencia: No le importa el nombre de la etiqueta, solo la posición física en la tabla.Slicing: Sigue la convención estándar de Python: el inicio es inclusivo y el final es exclusivo.
"""

df_sales.iloc[10:20, [0,3,2]]
```

```python
# FILTRAR Filtrado Básico (Boolean Indexing)
df_sales[df_sales['STATE'] == 'NY']
```

```python
# FILTAR 2
df_sales.query("STATE == 'NY' ")
```

```python
# # Uso de .loc (Recomendado)
# Es más explícito y te permite seleccionar tanto las filas (por condición) como las columnas específicas que quieres ver en un solo paso.
df_sales.loc[df_sales['STATE'] == 'NY', ['CUSTOMERNAME', 'STATE']]
```

```python
# FILTRAR COLUMNAS
df_sales.filter(like='ORDER')
```

```python
# Filtrado CONDICIONAL
df_sales[(df_sales.STATE == 'NY') & (df_sales.QTR_ID == 1)]
```

### EDA

```python
# Cliente por Pais
df_sales[['COUNTRY', 'CUSTOMERNAME']].groupby("COUNTRY").count()
```
