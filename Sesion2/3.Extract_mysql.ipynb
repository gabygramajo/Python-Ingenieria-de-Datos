```python
import sqlalchemy as db
import pandas as pd
```

```python
# 1. Crea el objeto 'engine' que define la configuraci칩n de la conexi칩n.
# Usa el driver 'mysqlconnector' para conectarse al usuario 'root' con contrase침a 'root'
# en el servidor local (localhost), puerto 3310 y apunta a la base de datos 'retail_db'.
engine = db.create_engine('mysql+mysqlconnector://root:root@localhost:3310/retail_db')

# 2. Establece y abre formalmente la conexi칩n activa con la base de datos.
# Este objeto 'conn' es el que se utiliza para ejecutar consultas (queries) SQL.
conn = engine.connect()
```

```python
# Extraer Datos de TODA LA TABLA
df_departments = pd.read_sql_table('departments', con=conn)
df_departments
```

```python
# Extraer Datos de mediante una query
df_departments = pd.read_sql_query("SELECT * FROM departments LIMIT 3", con=conn)
df_departments
```

```python
# Extraer Datos de mediante una query
df_departments = pd.read_sql("SELECT * FROM departments LIMIT 3", con=conn)
df_departments
```
